```python
def get_matrix_size():
    # Запрашиваем размерность, пока пользователь не введет корректное число
    while True:
        try:
            size_input = input("Введите размерность квадратной матрицы (целое число > 0): ")
            size = int(size_input)
            if size <= 0:
                print("Ошибка: размерность должна быть больше 0!")
                continue
            return size
        except ValueError:
            print("Ошибка: введите корректное целое число!")


def generate_random_number(seed_value):
    # Линейный конгруэнтный генератор псевдослучайных чисел (LCG)
    a, c, m = 1103515245, 12345, 2**31
    seed_value = (a * seed_value + c) % m
    return abs(seed_value % 100), seed_value


def create_matrix(size):
    # Создаем квадратную матрицу со случайными числами от 0 до 99
    matrix = []
    seed = 42  # Фиксированное начальное значение для воспроизводимости

    print("\nГенерация матрицы...")
    for i in range(size):
        row = []
        for j in range(size):
            random_num, seed = generate_random_number(seed)
            row.append(random_num)
        matrix.append(row)

    return matrix


def print_matrix(matrix):
    # Красивый вывод матрицы в табличном виде
    print("\nСгенерированная матрица:")
    print("-" * (4 * len(matrix) + 4))
    for row in matrix:
        formatted_row = " ".join(f"{elem:3d}" for elem in row)
        print(f"| {formatted_row} |")
    print("-" * (4 * len(matrix) + 4))


def calculate_diagonal_sums(matrix):
    # Считаем суммы элементов главной и побочной диагоналей
    size = len(matrix)
    main_diag_sum = 0
    secondary_diag_sum = 0
    main_diag_elements = []
    secondary_diag_elements = []

    for i in range(size):
        # Главная диагональ
        main_diag_elements.append(matrix[i][i])
        main_diag_sum += matrix[i][i]

        # Побочная диагональ
        secondary_diag_elements.append(matrix[i][size - 1 - i])
        secondary_diag_sum += matrix[i][size - 1 - i]

    # Для нечетной матрицы центральный элемент пересекается, вычитаем его один раз
    if size % 2 == 1:
        center = size // 2
        total_sum = main_diag_sum + secondary_diag_sum - matrix[center][center]
    else:
        total_sum = main_diag_sum + secondary_diag_sum

    return main_diag_sum, secondary_diag_sum, total_sum, main_diag_elements, secondary_diag_elements


def print_results(main_sum, secondary_sum, total_sum, main_elements, secondary_elements):
    # Выводим итоговые результаты с подробными пояснениями
    print("\n" + "=" * 50)
    print("РЕЗУЛЬТАТЫ ВЫЧИСЛЕНИЙ")
    print("=" * 50)

    print(f"\nГлавная диагональ: {main_elements}")
    print(f"-> Сумма главной диагонали: {main_sum}")

    print(f"\nПобочная диагональ: {secondary_elements}")
    print(f"-> Сумма побочной диагонали: {secondary_sum}")

    print(f"\n{'-' * 50}")
    print(f"ОБЩАЯ СУММА элементов обеих диагоналей: {total_sum}")
    print("=" * 50)


def main():
    # Точка входа в программу
    print("=" * 50)
    print("ПРОГРАММА: Сумма элементов диагоналей матрицы")
    print("=" * 50)

    size = get_matrix_size()
    print(f"\nРазмерность матрицы: {size}x{size}")

    matrix = create_matrix(size)
    print_matrix(matrix)

    results = calculate_diagonal_sums(matrix)
    print_results(*results)

    print("\nПрограмма успешно завершила работу!")


if __name__ == "__main__":
    main()
```
